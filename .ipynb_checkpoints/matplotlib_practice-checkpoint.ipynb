```python
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.plot(x_axis, y_axis)
#SHOW is ALWAYS last!! and use it once per session- otherwise itll cause errors
plt.show()
```

```python
#shows the same as above 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
#To change figure-level attributes (such as axis labels, a title, or a legend) 
#or save the figure as an image, use fig = plt.figure().
```

```python
#you could change the line color thickeness etc

plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label= 'Boston')
plt.xlabel("Date")
plt.ylabel("Fares($)")
#set a limit (range)
plt.ylim(0,45)
plt.title ("PyBer Fare by Month")
#add a grid
plt.grid()
#adding a legend (lil box)
plt.legend()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.bar(x_axis,y_axis)
```

```python
plt.bar(x_axis, y_axis, color="green", label= "boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#horizontal bar(switch the x and y)IMPORTANT!!! 
plt.barh(y_axis, x_axis)
```

```python
#.gca() puts the months in ascending order 
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

```python
fig, ax= plt.subplots()
ax.scatter(y_axis, x_axis, color= "skyblue", linewidth=2, s=[i*5 for i in y_axis], label= "Boston")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.pie(y_axis, labels= x_axis)
plt.show()
```

```python
colors= ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#why is .2 equal to july?
#adjust the size BEFORE you enter 
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode= explode_values,
        colors=colors,
        labels= x_axis, 
        autopct='%.1f%%')
plt.show()
```
